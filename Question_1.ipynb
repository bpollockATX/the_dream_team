```python
# importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gmaps
import scipy.stats as st
from statistics import mean 
import seaborn as sns
# from api_keys import g_key
```

#### Preparing our DataFrame for the Analysis

```python
# reading and displaying the cleaned input csv file

input_path = "cleaned_data.csv"

df = pd.read_csv(input_path)

# drop rows with all Nan in the row:

df = df.dropna(how = "all")
df.head()
```

```python
# Calculating Expenditure per student and adding to the dataframe

Q1_df = df.copy()
Q1_df["Total Expenditures"] = Q1_df["Total Expenditures"]*1000
Q1_df["Total Revenues"] = Q1_df["Total Revenues"]*1000
Q1_df["Expenditure per Student"] = df["Total Expenditures"]/df["Enrollment"]
Q1_df.head()
```

```python
# Adding latitude and longitude to the dataframe needed for heatmap

location_data = pd.read_csv("lat_lgt_data.csv")
Q1_df = pd.merge(location_data, Q1_df, on = "State", how = "outer")

Q1_df.head()
```

```python
# find and displaying years

number_of_year_coverage = len(Q1_df["Year"].value_counts())
min_year = min(df["Year"])
max_year = max(df["Year"])
year_coverage =[i for i in range(min_year,max_year)]

year_coverage_df = pd.DataFrame({"Year":year_coverage})
year_coverage_df
```

```python
## find and displaying the states

states = ["AL"]

number_of_states = Q1_df["State"].nunique()
print(f"Number of States is {number_of_states} including:")
for i in range(len(Q1_df["State"])-1):
    if Q1_df["State"][i+1]!=Q1_df["State"][i]:
        states.append (Q1_df["State"][i+1]) 

states_df = pd.DataFrame({"States": states})
states_df 
```

### Heat Map to see which states have the most expenditure per State

```python
# grouped by state and get the average 

Q1_grouped_df = Q1_df.groupby("State").mean().reset_index()
Q1_grouped_df = Q1_grouped_df.dropna()
 

locations = [(i,j) for i,j in zip(Q1_grouped_df.loc[:,"Latitude"], 
            Q1_grouped_df.loc[:, "Longitude"])]

weight = Q1_grouped_df.loc[:,"Expenditure per Student"]
    

# Customize the size of the figure

figure_layout = {
    'width': '800px',
    'height': '500px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}


fig = gmaps.figure(layout = figure_layout)
max_intensity = max(weight)
             
# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights = weight, 
                                 dissipating = False, max_intensity = max_intensity,
                                 point_radius= 0.75)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

#### Based on the info in the heat map, the states in the east coast of the United states and Alaska have the highest expenditure per student. 

```python
# Assign expenditure and test score data 


states = Q1_grouped_df.loc[:,"State"]
average_expenditure_per_state = Q1_grouped_df.loc[:,"Expenditure per Student"]
average_math_4th_per_state = Q1_grouped_df.loc[:,"Average Math Score (4th Grade)"]
average_math_8th_per_state = Q1_grouped_df.loc[:,"Average Math Score (8th Grade)"]
average_reading_4th_per_state = Q1_grouped_df.loc[:,"Average Reading Score (4th Grade)"]
average_reading_8th_per_state = Q1_grouped_df.loc[:,"Average Reading Score (8th Grade)"]
```

```python
# Assign expenditure and test score data of each state to a dictionary with states as key values

state_per_year_original = {}
average_expenditure_per_year= {}
average_math_4th_per_year = {}
average_math_8th_per_year = {}
average_reading_4th_per_year = {}
average_reading_8th_per_year = {}

for state in states:
    
        average_expenditure_per_year [state] = Q1_df.loc[Q1_df["State"]==state,"Expenditure per Student"].values.tolist()
        average_math_4th_per_year [state] = Q1_df.loc[Q1_df["State"]==state,"Average Math Score (4th Grade)"].values.tolist()
        average_math_8th_per_year [state] = Q1_df.loc[Q1_df["State"]==state,"Average Math Score (8th Grade)"].values.tolist()
        average_reading_4th_per_year [state] = Q1_df.loc[Q1_df["State"]==state,"Average Reading Score (4th Grade)"].values.tolist()
        average_reading_8th_per_year [state] = Q1_df.loc[Q1_df["State"]==state,"Average Reading Score (8th Grade)"].values.tolist()
```

```python
# Expenditure per Student
############################################################################################################################

# Change the style of figure
sns.set()

# geting x and y values for the bar graph

x_values = states
y_values = average_expenditure_per_state

# sort the states from the max expenditure to the least expenditure:

y_sorted = sorted(y_values, reverse = True )
x_sorted = [x_val for _,x_val in sorted(zip(y_values,x_values), reverse = True)]

fig_average_expenditure, ax = plt.subplots(figsize = (20,10))
fig_average_expenditure.suptitle("Average Expenditure per Pupil vs. State", y= 0.95, fontsize = 25)

for i in range(len(x_sorted)):
    if i<=4:
        plt.bar(x_sorted[i], y_sorted[i], width = 0.4, color = "blue", alpha =0.5) 
        fontweight='bold'
    elif i>=45:
        plt.bar(x_sorted[i], y_sorted[i], width = 0.4, color = "red", alpha = 0.5)
    else:
        plt.bar(x_sorted[i], y_sorted[i], width = 0.4, color = "white")
        
plt.ylabel("Average Expenditure per Pupil ($)", fontsize = 22)
plt.xticks(x_sorted, rotation=90) 
plt.tick_params(axis="x", labelsize = 15)

# Save and display the bar chart

plt.savefig('Q1_Images/expenditure_per_student_vs_state.png')
plt.show()
```

##### Based on the above chart, the top 5 states based on the expenditure per students (green bars) have a distribution of test scores from high to low.

##### Also, the bottom 5 states based on the expenditure per students (red bars) have a distribution of test scores from middle to low. 

##### Limitations:

1- There are many interrelated factors that can affect the performance of students of a state; therefore, evaluating the effect of expenditure per student as an independent factor on the students' test score does not give a comprehensive vision

2- Our data was not included expenditure per student for each district in a state. In dense populated states like New York and District of Columbia with a very diverse school budget, the average expenditure of the state doesnt represent the average expenditure of the districts.

3- While our test scores data was only included the math and reading score of 4th and 8th grade from 1995-2016, it is a good random sample to evaluate teh performance of students between 4th and 8th grade. 

##### Conclusions:

1- The states (green bars) with the highest expenditure per student have shown a different test score patterns. The test results are distributed betwean low to high; therefore, our hypothesis is that there is a sweet point in the average expenditure in which the best score tests appear. 

2- The test score of the states (red bars) with the lowest expenditure per student are distributed betwean middle to low. In other words, none of the states with a low school budget has shown a hight performance in standard tests (math and reading). This conclusion make us more confident in the correctness of our hypothesis
